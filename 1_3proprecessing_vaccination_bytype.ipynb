```python
import pandas as pd
```

```python
pd_data = pd.read_csv('./datas/vaccination_bytype_20210719.csv',encoding='euc-kr',thousands=',')
```

```python
type(pd_data.iloc[0:,3])
len(pd_data.iloc[0:,3])
type(pd_data.iloc[142,3])
```

```python
azones =[];
aztwos =[];
hzones =[];
hztwos =[];
yss =[];
mdones =[];
mdtwos=[];

for index in range(0,142):
    azones.append(pd_data.iloc[index,3]-pd_data.iloc[index+1,3])
    aztwos.append(pd_data.iloc[index,4]-pd_data.iloc[index+1,4])
    hzones.append(pd_data.iloc[index,5]-pd_data.iloc[index+1,5])
    hztwos.append(pd_data.iloc[index,6]-pd_data.iloc[index+1,6])
    yss.append(pd_data.iloc[index,7]-pd_data.iloc[index+1,7])
    mdones.append(pd_data.iloc[index,8]-pd_data.iloc[index+1,8])
    mdtwos.append(pd_data.iloc[index,9]-pd_data.iloc[index+1,9])

azones.append(0)
aztwos.append(0)
hzones.append(0)
hztwos.append(0)
yss.append(0)
mdones.append(0)
mdtwos.append(0)
```

```python
pd_data['아스트라제네카1차 일일접종량'] =azones
pd_data['아스트라제네카 완료 일일접종량'] =aztwos
pd_data['화이자 1차 일일접종량'] =hzones
pd_data['화이자 완료 일일접종량'] =hztwos
pd_data['얀센 완료 일일접종량'] =yss
pd_data['모더나 1차 일일접종량'] =mdones
pd_data['모더나 완료 일일접종량'] =mdtwos
```

```python
az_pd = pd_data[['일자','아스트라제네카 1차','아스트라제네카 완료\n(AZ-PF교차포함)','아스트라제네카1차 일일접종량','아스트라제네카 완료 일일접종량']]
az_pd['종류'] = '아스트라제네카'

hz_pd = pd_data[['일자','화이자 1차','화이자 완료\n(AZ-PF교차미포함)','화이자 1차 일일접종량','화이자 완료 일일접종량']]
hz_pd['종류'] = '화이자'

md_pd = pd_data[['일자','모더나 1차','모더나 완료','모더나 1차 일일접종량','모더나 완료 일일접종량']]
md_pd['종류'] = '모더나'

ys_pd = pd_data[['일자','얀센 1차(완료)','얀센 완료 일일접종량']]
ys_pd['종류'] = '얀센'
ys_pd['백신 1차 누적 접종량'] =0
ys_pd['백신 1차 일일 접종량']=0
```

```python
az_pd.rename(columns={'아스트라제네카 1차':'백신 1차 누적 접종량','아스트라제네카 완료\n(AZ-PF교차포함)':'백신 완료 누적 접종량','아스트라제네카1차 일일접종량':'백신 1차 일일 접종량','아스트라제네카 완료 일일접종량':'백신 완료 일일 접종량'},inplace=True)
hz_pd.rename(columns={'화이자 1차':'백신 1차 누적 접종량','화이자 완료\n(AZ-PF교차미포함)':'백신 완료 누적 접종량','화이자 1차 일일접종량':'백신 1차 일일 접종량','화이자 완료 일일접종량':'백신 완료 일일 접종량'},inplace=True)
md_pd.rename(columns={'모더나 1차':'백신 1차 누적 접종량','모더나 완료':'백신 완료 누적 접종량','모더나 1차 일일접종량':'백신 1차 일일 접종량','모더나 완료 일일접종량':'백신 완료 일일 접종량'},inplace=True)
ys_pd.rename(columns={'얀센 1차(완료)':'백신 완료 누적 접종량','얀센 완료 일일접종량':'백신 완료 일일 접종량'},inplace=True)
```
